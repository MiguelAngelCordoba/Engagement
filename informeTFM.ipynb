```python
df3 = df2.copy()
```

```python
# Restar 1 a todas las filas en las columnas 15 a 31 (UWES)
df3.iloc[:, 15:32] = df3.iloc[:, 15:32] - 1

vigor = df3.iloc[:, [15, 18, 22, 26, 29,  31]]

# Calcular el promedio de cada columna
media_vigor = vigor.mean().mean()
media_vigor = media_vigor.round(2)

dedicacion = df3.iloc[:, [16, 19, 21, 24, 27]]

# Calcular el promedio de cada columna
media_dedicacion = dedicacion.mean().mean()
media_dedicacion = media_dedicacion.round(2)

absorcion = df3.iloc[:, [17, 20, 23, 25, 28, 30]]

# Calcular el promedio de cada columna
media_absorcion = absorcion.mean().mean()
media_absorcion = media_absorcion.round(2)

Total = ((media_vigor + media_dedicacion + media_absorcion)/3)
Total = Total.round(2)
```

```python
# Agregamos columnas con promedio

# Calcular el promedio de las columnas especificadas y asignar a una nueva columna
df3['PromedioVigor'] = df3.iloc[:, [15, 18, 22, 26, 29,  31]].mean(axis=1)  # Usar índices 0-based
df3['PromedioVigor'] = df3['PromedioVigor'].round(2)

df3['PromedioDedicacion'] = df3.iloc[:, [16, 19, 21, 24, 27]].mean(axis=1)  # Usar índices 0-based
df3['PromedioDedicacion'] = df3['PromedioDedicacion'].round(2)

df3['PromedioAbsorcion'] = df3.iloc[:, [17, 20, 23, 25, 28, 30]].mean(axis=1)  # Usar índices 0-based
df3['PromedioAbsorcion'] = df3['PromedioAbsorcion'].round(2)
```
